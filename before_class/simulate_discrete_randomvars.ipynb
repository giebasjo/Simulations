## Description: 

The `make_csv.py` file creates the csv file containing the information of the probability mass function. See the description for more information on formatting and such. <br>
The `alias_table_construction.py` file builds the alias table needed for simulating discrete random variables from the information obtained through the csv generated by `make_csv.py`. For more information, see the description. <br>
This file, uses the alias table generated by `alias_table_construction.py` to simulate a finite, disctete probability distribution. The following algorithm was presented in the **MSCF: Simulations for Option Pricing** course.
